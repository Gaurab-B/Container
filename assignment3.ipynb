```python
import pandas as pd
```

```python
path = "https://github.com/Gaurab-B/Crime_Rate_Prediciton/blob/main/census.csv?raw=true"
census = pd.read_csv(path,engine='python',encoding='latin1')
census
```

```python
modified_census =census.loc[:, ["STNAME","ESTIMATESBASE2010",	"POPESTIMATE2010",	"POPESTIMATE2011",	"POPESTIMATE2012",	"POPESTIMATE2013",	"POPESTIMATE2014",	"POPESTIMATE2015"]]
modified_census.tail(5)
```

```python
modified_census["sum"] = modified_census.ESTIMATESBASE2010 + modified_census.POPESTIMATE2010 + modified_census.POPESTIMATE2011 + modified_census.POPESTIMATE2012 + modified_census.POPESTIMATE2013 + modified_census.POPESTIMATE2014 + modified_census.POPESTIMATE2015
modified_census.tail(5)
```

```python
valid_columns = ['GR2010to2011', 'GR2011to2012', 'GR2013to2014', 'GR2014to2015']
```

```python
def census_calc(growth_col_name,current_col, prev_col ):
    modified_census[growth_col_name] = ((modified_census[current_col] - modified_census[prev_col]) / modified_census[prev_col]) * 100
    modified_census[growth_col_name]
```

```python
census_calc("GR2010to2011","POPESTIMATE2011","POPESTIMATE2010")
census_calc("GR2011to2012","POPESTIMATE2012","POPESTIMATE2011")
census_calc("GR2013to2014","POPESTIMATE2014","POPESTIMATE2013")
census_calc("GR2014to2015","POPESTIMATE2015","POPESTIMATE2014")
```

```python
modified_census.tail(5)
```

```python
modified_census["AverageGrowthRate"] = modified_census[valid_columns].mean(axis=1)
modified_census.tail(5)
```

```python
positive_predictions = modified_census[modified_census["AverageGrowthRate"] > 0]
positive_predictions.tail(5)
```

```python
#positive_predictions["PopPredict2016"] = positive_predictions["POPESTIMATE2015"]+   positive_predictions["AverageGrowthRate"] * positive_predictions["POPESTIMATE2015"]
positive_predictions.loc[:, "PopPredict2016"] = positive_predictions["POPESTIMATE2015"] + positive_predictions["AverageGrowthRate"] * positive_predictions["POPESTIMATE2015"]
positive_predictions
```
